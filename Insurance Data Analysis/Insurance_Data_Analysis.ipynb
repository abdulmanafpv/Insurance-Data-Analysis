BMI of womens with children and no children

d) Is the distribution of BMI across women with no children, one child and two children, the same ?

```python
plt.figure(figsize=(10,7))
ax=plt.axes(projection='3d')
z=data['female']
y=data['bmi']
x=data['children']
plt.xlabel('children')
plt.ylabel('bmi')
ax.set_zlabel('female')
plt.title('BMI of womens with childrens ')
ax.plot3D(x,y,'red')
plt.show()
```

From this graph we can understand the women who have 0,1,2,3 childrens have more BMI when compared to 4 and 5 children womens. Also we can note down that the womens who have 0,1,2 children have more BMI when compared to 3, 4 and 5 children womens. So we can say that when children increases to a women it affect their BMI. in 0,1,2 children womens have not much changes in their BMI. when their childrens increase more than two, their BMI become decreasing.
